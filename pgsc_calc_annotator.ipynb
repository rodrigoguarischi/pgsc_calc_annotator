```python
# Import all necessary libraries

import pandas as pd
import numpy as np
import os
from xlsxwriter.utility import xl_rowcol_to_cell

# Set working directory
os.chdir("/Users/rgsousa/Desktop/pgsc_calc_annotator/")
```

```python
### Read all tabular files (scores, ancestry, and PGS metadata and merge accordingly 

## Read raw and adjusted PGS scores
pgsc_calc_scores = pd.read_table(
    "./input_data/gss-full-105-1000G-HGDP_pgs.txt",
    sep = "\t"
    )
display( pgsc_calc_scores )



## Load ancestry information
pgsc_calc_ancestries = pd.read_table(
    "./input_data/gss-full-105-1000G-HGDP_popsimilarity.txt",
    sep = "\t"
    )
# Filter down to some columns
selected_columns = ["IID", "MostSimilarPop", "MostSimilarPop_LowConfidence", "REFERENCE", "SuperPop"]
# selected_columns_full = ["sampleset", "IID", "RF_P_AFR", "RF_P_AMR", "RF_P_CSA", "RF_P_EAS", "RF_P_EUR", "RF_P_MID", "MostSimilarPop", "MostSimilarPop_LowConfidence", "REFERENCE", "SuperPop"]
pgsc_calc_ancestries_subset = pgsc_calc_ancestries[ selected_columns ]
display( pgsc_calc_ancestries_subset )



## Load PGS metadata
pgsc_calc_info = pd.read_table(
    "./input_data/pgs_all_metadata_scores.csv",
    sep = ",",
    dtype = "str"
    )
display( pgsc_calc_info )
```

```python
### Aggregate all dataframes into one called pgsc_calc_merged and make some adjustments

## Merge scores and ancestry info
pgsc_calc_merged = pd.merge(
    left = pgsc_calc_scores,
    right = pgsc_calc_ancestries_subset,
    how = "left",
    left_on = "IID",
    right_on = "IID"
    )
# Drop the REFERENCE samples and drop some unnecessary columns
pgsc_calc_merged = pgsc_calc_merged[ pgsc_calc_merged["REFERENCE"] == False ].drop(
    labels = ["REFERENCE", "SuperPop"],
    axis = 1
    )
# display( pgsc_calc_merged )

## Merge the new dataframe with PGS metadata
# Create a columns to be used as key in the merge
pgsc_calc_merged["PGS_ID"] = pgsc_calc_merged["PGS"].str.split( "_" ).str[0]
pgsc_calc_merged = pd.merge(
    left = pgsc_calc_merged,
    right = pgsc_calc_info,
    left_on = "PGS_ID",
    right_on = "Polygenic Score (PGS) ID",
    )
# Add a column that is a hyperlink to the pgs_catalog
pgsc_calc_merged["pgs_catalog_hyperlink"] = "https://www.pgscatalog.org/score/" + pgsc_calc_merged["PGS_ID"].astype( str ) + "/"

# Adjust column percentile_MostSimilarPop to float
pgsc_calc_merged["percentile_MostSimilarPop"] = pgsc_calc_merged["percentile_MostSimilarPop"] / 100

# Order result table by percentile_MostSimilarPop
pgsc_calc_merged = pgsc_calc_merged.sort_values(
    by = "percentile_MostSimilarPop",
    ascending = False
    )
display( pgsc_calc_merged )
```

Write output to an **excel** file

```python
# Export dataframe into a xls file
writer = pd.ExcelWriter(
    path = "./gss_pgs_annotated.xls",
    engine = 'xlsxwriter'
    )

pgsc_calc_merged.to_excel(
    writer,
    sheet_name='gss_pgs_summary',
    index = False
    )

# Get access to the workbook and sheet
workbook = writer.book
worksheet = writer.sheets['gss_pgs_summary']

# Automatically add filters to columns
worksheet.autofilter( "A1:AF{0}".format( len( pgsc_calc_merged.index ) + 1) )

# Add a left-alignment format
left_alignment_fmt = workbook.add_format({'align': 'left'})

# Add a percent format with 2 decimal points
percent_fmt = workbook.add_format({'num_format': '0.00%', 'align': 'left'})

# Format the columns by width and include number formats
worksheet.set_column('A:G', 20, left_alignment_fmt)         # 
worksheet.set_column('H:H', 20, percent_fmt)                # percentile_MostSimilarPop
worksheet.set_column('I:AG', 20, left_alignment_fmt)        # 

# Create a red, yellow and green fills with dark text
red_format = workbook.add_format({
    'bg_color': '#FFC7CE',
    'font_color': '#9C0006'
    })
yellow_format = workbook.add_format({
    'bg_color': '#FFEB9C',
    'font_color': '#9C6500'
    })

# Create confitional formating (>95%)
cells_location = 'H2:H' + str( len(pgsc_calc_merged) + 1)
worksheet.conditional_format(
    cells_location,
    {'type': 'cell',
     'criteria': '>=',
     'value': '0.95',
     'format': red_format}
     )

# Create confitional formating (< 5%)
worksheet.conditional_format(
    cells_location,
    {'type': 'cell',
     'criteria': '<=',
     'value': '0.05',
     'format': yellow_format}
     )

# Save modifications
writer.close()
```
